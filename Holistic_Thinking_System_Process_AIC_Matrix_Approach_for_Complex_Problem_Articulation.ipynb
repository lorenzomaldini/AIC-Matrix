# Introduction

The process is built upon existing Systems ontologies of AIC (by Dr William Smith) and Institutional Compass (by Prof Michelle Friend) Systems Theories. The architecture of the process itself follows the Cynefin Framework as interpreted by Gary Smith and Colleagues (see references section).

to perform AIC-Matrix approach, we need some enabling general assumptions:

assumption 1) for every purpose and action there is a tacit intent.

assumption 2) intent of a purpose can be in the contetx of: Appreciation (apr), Influence (inf), Control (ctr).

assumption 3 the compostion of a puprose and action is denoted as complex behaviour.

assumption 4) when an acting compelx behaviour is engaged with a reacting complex behaviour, it resolves into an emergent effect.

assumption 5) there are four possible resultant contexts a behaviour's effect can be in:

    5.1: a complex behaviour denies another complex behaviour from manifesting its purpose using its capabilty.
    
    5.2: a complex behaviour reinforces another complex beaviour to manifest its purpose using its capabilty.
    
    5.3: a complex behaviour circumvents the possible effect of another complex behaviour on its capabilty to manifest its purpose.
    
    5.4: a complex behaviour ignores the possible effect of another complex behaviour on its capabilty to manifest its purpose.

the tuple of (source_purpose_effect_type, source_capability_effect_type, sink_purpose_effect_type, sink_capability_reaction_effect_type) describes a possible "Context or Complicatedness" of a complex behaviour may result into, for example (deny, demy, ignore, cricumvent)

assumption 6) a complex behaviour can have a hoilistic pattern of attitude.

assumption 7) there are 3 holistic patterns a complex behaviour can be described being in: Harmony, Discipline, Excitement.

assumption 8) the holistic attitude of a complex behaviour depends on the contribution of types of effects in a given context of complicatedness: 

    deny = {5% harmony, 5% excitement, 90% dicipline}
    
    reinforce = {90% harmony, 5% excitement, 5% dicipline}
    
    circumvent = {30% harmony, 40% excitement, 30% dicipline}
    
    ignore = {20% harmony, 60% excitement, 20% dicipline}

See step 2.4.1 for justification behind the chosen values. At this stage of the reseach, the justification is subjective but in teh future we expect more rigor in figuring out how to evalue H,E,D values. The Institutional Compass has its own methodology which we will examin its applicabilty.

## Step 2.3) Identify as-is Complex Behaviours' Possible Conexts of Complictaednessof 

Complicatedness Contexts are scenarios where interacting source and sink may possibly find themselves in. They are derivable by fixing the purpose and capability types for each component behavior and changing the values of (source_purpose_effect_type, source_capability_effect_type, sink_purpose_effect_type, sink_capability_reaction_effect_type) between {deny, reinforce, circumvent, ignore}. for example:

#### (Deny, Deny, Deny, Deny) Complicatedness Context:
Eagle-Robot (Source): Controlling purpose to inhibit adversarial drone behavior (deny effect) using appreciative capability to monitor (deny effect) the adversarial drone behavior.
Adversarial Drone (Sink): Influential purpose to disrupt policing Eagle-Robot behavior (deny effect) using influential capability of evading Eagle-Robot's effect (deny effect) on itself.

Scenario example: An Eagle-Robot is deployed to protect the train track zone from an adversarial drone. Both the Eagle-Robot and the adversarial drone engage in a game of cat and mouse (having clear purposes on each other), trying to deny each other's actions. The Eagle-Robot tries to intercept and disable the adversarial drone, while the adversarial drone tries to disrupt the Eagle-Robot's operations by jamming its communication systems.

#### (Deny, Reinforce, Deny, Reinforce) Complicatedness Context:
Eagle-Robot (Source): Controlling purpose to inhibit adversarial drone behavior (deny effect) using appreciative capability to monitor (reinforce effect) the adversarial drone behavior.
Adversarial Drone (Sink): Influential purpose to disrupt policing Eagle-Robot behavior (deny effect) using influential capability of evading Eagle-Robot's effect on itself (reinforce effect).

Scenario example: The Eagle-Robot attempts to stop the adversarial drone from entering the train track zone. However, the adversarial drone (purposefully) takes advantage to reinforce the Eagle-Robot's focus on it and lures the Eagle-Robot away from the train track zone, allowing other adversarial drones to enter the area undetected.

#### (Deny, Circumvent, Deny, Circumvent) Complicatedness Context:
Eagle-Robot (Source): Controlling purpose to inhibit adversarial drone behavior (deny effect) using appreciative capability to monitor the adversarial drone behavior (circumvent effect).
Adversarial Drone (Sink): Influential purpose to disrupt policing Eagle-Robot behavior (circumvent effect) using influential capability of evading Eagle-Robot's effect on itself (deny effect).

Scenario example: The Eagle-Robot tries to prevent the adversarial drone from entering the train track zone. The adversarial drone, instead of directly confronting the Eagle-Robot, (purposefully) uses its evasive capabilities to avoid detection and bypass the Eagle-Robot's surveillance, thus managing to enter the train track zone without being detected.

## Step 2.4.2) Perform Complex Behaviour Deficiency analysis for each context

When we consider deficiency, we look at how lacking a complex behavior as a whole is in terms of having the interacting components (source and sink) increase the chances for their respective purposes to be manifested.

#### Complex Behaviour Deficiency performance in Deny-Deny context

If the acting source behavior denies the reacting sink behavior, while the reacting sink behavior denies the acting source behavior, the overall complex behavior will be deficient (grade C). This is because both components have the purpose and capability to exert "denying" behavior over each other, leading to a stalemate where neither component can manifest their respective purposes effectively. In this situation, both components work against each other, and the overall complex behavior becomes ineffective.

Example: An Eagle-Robot is tasked with preventing an adversarial drone from disrupting the train network. The adversarial drone, on the other hand, is trying to deny the Eagle-Robot's ability to protect the network. In this scenario, both components work against each other, leading to a deficient complex behavior where neither the Eagle-Robot nor the adversarial drone can achieve their goals effectively.

#### Complex Behaviour Deficiency performance in Reinforce-Reinforce context

If the acting source behavior reinforces the reacting sink behavior, and the reacting sink behavior reinforces the acting source behavior, the overall complex behavior is said to be negligibly deficient (grade A). This is because both components have the purpose and capability to exert "cooperating" behavior, leading to a synergistic relationship where both components have more chances for their purposes to be manifested. In this situation, both components work together, and the overall complex behavior becomes highly effective.

Example: An Eagle-Robot is tasked with monitoring a train track zone to prevent trespassers from entering, while the police are responsible for apprehending any trespassers detected. In this scenario, the Eagle-Robot reinforces the police's ability to apprehend trespassers by providing accurate and timely information, while the police reinforce the Eagle-Robot's surveillance capabilities by acting on the information provided. This results in a negligibly deficient complex behavior, as both components effectively work together to achieve their goals.

#### Complex Behaviour Deficiency performance in Circumvent-Circumvent context

When the acting source behavior circumvents the reacting sink behavior, and the reacting sink behavior circumvents the acting source behavior, the overall complex behavior can be considered possibly deficient (grade B). In this case, both components are avoiding and managing the effects of the other component, but they lack the purpose and capability to directly influence or control each other. Their purposes might be partially manifested, but the complex behavior is not fully effective.

Example: An adversarial drone tries to enter a train track zone, and the policing Eagle-Robot is responsible for preventing unauthorized access. The drone tries to circumvent the Eagle-Robot's surveillance by taking evasive routes, while the Eagle-Robot adapts its monitoring strategy to track the drone. In this scenario, both components are trying to avoid the other's influence but are not directly denying or reinforcing each other. The overall complex behavior is possibly deficient as neither purpose is fully manifested.

#### Complex Behaviour Deficiency performance in Ignore-Ignore context

When the acting source behavior ignores the reacting sink behavior, and the reacting sink behavior ignores the acting source behavior, the overall complex behavior can also be considered possibly deficient (grade B). In this case, both components do not have a direct influence on each other, and their purposes might be partially manifested. The complex behavior is not fully effective because there is no strong relationship or interaction between the components.

Example: An Eagle-Robot is monitoring a train track zone to prevent trespassers from entering, while an adversarial drone is focused on a completely different mission, unrelated to the train track zone. Both components are not interacting or influencing each other, and their purposes are partially manifested. In this case, the Eagle-Robot's purpose of preventing trespassers might still be effective, but the complex behavior as a whole remains possibly deficient due to the lack of interaction between the components.

### General rules for Complex Behaviour Deficiency performance can be summarized as follows:

Rule 1) Same types of effects (Deny-Deny, Reinforce-Reinforce, Circumvent-Circumvent, Ignore-Ignore):

Complicatedness Context Example:

c. Circumvent-Circumvent: Possibly deficient (Grade B) - Both components avoid and manage each other's effects but lack the capability to directly influence or control each other, leading to partial effectiveness.

Complicatedness Context Example:

An adversarial drone tries to enter the train track zone, and the policing Eagle-Robot attempts to prevent unauthorized access. Both the drone and the Eagle-Robot try to avoid each other's influence, leading to a possibly deficient complex behavior where neither purpose is fully manifested.

This leads to a possibly deficient complex behavior where the full potential of cooperation is not realized, and overall effectiveness is compromised.

### How to evaluate Complex Behaviour Deficiency performance in some Complicatedness Context 

To compute the deficiency grade of a complex behaviour, we can use the following methodology:
step 1) Define the context of the complex behaviour in terms of the source and sink components' purpose and capability effect types Captured in a Complicatedness Context Factors: (source_purpose_effect_type, source_capability_effect_type, sink_purpose_effect_type, sink_capability_reaction_effect_type). 

step 2) Assign the corresponding harmony, excitement, and discipline percentages based on the given definitions of deny, reinforce, circumvent, and ignore. given that: 

    Deny = {5% harmony, 5% excitement, 90% discipline},
    Reinforce = {90% harmony, 5% excitement, 5% dicipline}, 
    Circumvent = {30% harmony, 40% excitement, 30% discipline}, 
    Ignore = {20% harmony, 60% excitement, 20% discipline}


step 3) Calculate the average harmony, excitement, and discipline percentages for the complex behavior:

    o Average Harmony: (source_purpose_effect_type_harmony + source_capability_effect_type_harmony + sink_purpose_effect_type_harmony + sink_capability_reaction_effect_type_harmony) / 4
    
    o Average Excitement: (source_purpose_effect_type_excitement + source_capability_effect_type_excitement + sink_purpose_effect_type_excitement + sink_capability_reaction_effect_type_excitement) / 4
    
    o Average Discipline: (source_purpose_effect_type_discipline + source_capability_effect_type_discipline + sink_purpose_effect_type_discipline + sink_capability_reaction_effect_type_discipline) / 4
    
step 4) Determine the deficiency grade (Grade A, Grade B, Grade C) based on the calculated average harmony, excitement, and discipline percentages:

    o Grade A (Negligibly Deficient): discipline < excitement + harmony 
    
    o Grade B (Possibly Deficient): if not (discipline < excitement + harmony) or discipline > excitement + harmony). If Discipline > harmony, then it is an “Grade B leaning towards Grade C” , if harmony > Discipline , then it is an “Grade B leaning towards Grade A” 
    
    o Grade C (Deficient): discipline > excitement + harmony
    
step 5) Assign the deficiency grade to the complex behaviour based on the calculated averages.

Here is an example:
Consider a complex behavior with the following context:

    •	source_purpose_effect_type = deny

    •	source_capability_effect_type = reinforce

    •	sink_purpose_effect_type = circumvent

    •	sink_capability_reaction_effect_type = ignore
Using the given definitions, we assign the harmony, excitement, and discipline percentages:

    •	deny: {5% harmony, 5% excitement, 90% discipline}

    •	reinforce: {90% harmony, 5% excitement, 5% discipline}

    •	circumvent: {30% harmony, 40% excitement, 30% discipline}

    •	ignore: {20% harmony, 60% excitement, 20% discipline}
Calculate the average harmony, excitement, and discipline percentages:

    •	Average Harmony: (5 + 90 + 30 + 20) / 4 = 36.25%

    •	Average Excitement: (5 + 5 + 40 + 60) / 4 = 27.5%

    •	Average Discipline: (90 + 5 + 30 + 20) / 4 = 36.25%
    
Based on the averages, this complex behaviour corresponds to Grade B (Possibly Deficient) deficiency performance, because it is neither discipline < excitement + harmony , nor discipline > excitement + discipline. it is possibly deficient that could potentially lean towards Harmony or Discipline since harmony = discipline


```python
# Define the deficiency for each effect type
attitudes = {
    "deny": {"harmony": 0.05, "excitement": 0.05, "discipline": 0.90},
    "reinforce": {"harmony": 0.90, "excitement": 0.05, "discipline": 0.05},
    "circumvent": {"harmony": 0.30, "excitement": 0.40, "discipline": 0.30},
    "ignore": {"harmony": 0.20, "excitement": 0.60, "discipline": 0.20},
    "unclear": {"harmony": 0.33, "excitement": 0.33, "discipline": 0.33},
}

# Function to compute deficiency performance grade
def compute_deficiency_performance_grade(row):
    source_purpose_effect_type = row["source purpose effect type"]
    source_capability_effect_type = row["source capability effect type"]
    sink_purpose_effect_type = row["sink's purpose effect type"]
    sink_capability_reaction_effect_type = row["sink's capability reaction effect type"]
    
    effect_types = [
        source_purpose_effect_type,
        source_capability_effect_type,
        sink_purpose_effect_type,
        sink_capability_reaction_effect_type,
    ]

    total_harmony = 0
    total_excitement = 0
    total_discipline = 0

    for effect_type in effect_types:
        total_harmony += attitudes[effect_type]["harmony"]
        total_excitement += attitudes[effect_type]["excitement"]
        total_discipline += attitudes[effect_type]["discipline"]

    avg_harmony = total_harmony / len(effect_types)
    avg_excitement = total_excitement / len(effect_types)
    avg_discipline = total_discipline / len(effect_types)

    
    if avg_harmony > avg_excitement + avg_discipline:
        grade = "A"#  grade a deficiency
    elif avg_discipline > avg_harmony + avg_excitement:
        grade = "C" #  grade c deficiency
    else:
        grade = "B"

    return grade

# Assuming df_combinations is a DataFrame with the required columns
df_combinations["deficiency performance grade"] = df_combinations.apply(compute_deficiency_performance_grade, axis=1)
```

```python
def create_matrix_image(base_matrix, components):
    N = len(components) + 1
    img_size = 1028
    cell_size = img_size // N
    
    image = Image.new("RGB", (img_size, img_size), "white")
    draw = ImageDraw.Draw(image)
    font = ImageFont.truetype("arial.ttf", 27)
    bold_font = ImageFont.truetype("arialbd.ttf", 90)  # Bold font for deficiency grade
    
       
    # Draw the problem_standalone_components names
    for i, component in enumerate(components):
        draw.text((cell_size * (i + 1) + cell_size // 2 - 5 * len(component), cell_size // 2 - 10), component, font=font, fill="black")
        draw.text((cell_size // 2 - 5 * len(component), cell_size * (i + 1) + cell_size // 2 - 10), component, font=font, fill="black")
    
    # Draw the matrix
    for row, col in itertools.product(components, repeat=2):
        i = components.index(row)
        j = components.index(col)
        
        pair = next((p for p in base_matrix if p["source"] == row and p["sink"] == col), {})
        source_purpose_type = pair.get("source purpose type", "")
        
        if source_purpose_type == "inf":
            color = "red"
        elif source_purpose_type == "apr":
            color = "green"
        elif source_purpose_type == "ctr":
            color = "blue"
        elif source_purpose_type == "unclear":
            color = "white"
        else:
            color = "black"
        
        draw.rectangle([(cell_size * (j + 1), cell_size * (i + 1)), (cell_size * (j + 2), cell_size * (i + 2))], fill=color)

        # Draw the attitude_box
        attitude_tuple = pair.get("attitude_tuple (h,e,d)", ())
        if not attitude_tuple or len(attitude_tuple) != 3:
            attitude_color = (0, 0, 0)  # default color
        else:
            h, e, d = attitude_tuple
            attitude_color = (int(255 * h), int(255 * e), int(255 * d))
            print(f"attitude_color = {attitude_color}")
        
        
        attitude_box_size = cell_size // 2
        attitude_box_x1 = cell_size * (j + 1) + (cell_size - attitude_box_size) // 2
        attitude_box_y1 = cell_size * (i + 1) + (cell_size - attitude_box_size) // 2
        attitude_box_x2 = attitude_box_x1 + attitude_box_size
        attitude_box_y2 = attitude_box_y1 + attitude_box_size

        draw.rectangle([(attitude_box_x1, attitude_box_y1), (attitude_box_x2, attitude_box_y2)], fill=attitude_color)
    
        # Draw the deficiency grade
        deficiency_grade = pair.get("deficiency performance grade", "")
        if deficiency_grade:
            # Calculate the opposite color of the attitude_box
            opposite_color = tuple(map(lambda x: 255 - x, attitude_color))
            # Draw the deficiency grade letter in the center of the attitude_box
            deficiency_grade_x = attitude_box_x1 + (attitude_box_size // 2) - (bold_font.getsize(deficiency_grade)[0] // 2)
            deficiency_grade_y = attitude_box_y1 + (attitude_box_size // 2) - (bold_font.getsize(deficiency_grade)[1] // 2)
            draw.text((deficiency_grade_x, deficiency_grade_y), deficiency_grade, font=bold_font, fill=opposite_color)
        # Draw the black borders
        for i in range(N + 1):
            draw.line([(i * cell_size, 0), (i * cell_size, img_size)], fill="black", width=1)
            draw.line([(0, i * cell_size), (img_size, i * cell_size)], fill="black", width=1)

    return image

# clean standalone_components
def clean_list(input_list):
    cleaned_list = []
    
    for item in input_list:
        cleaned_item = item.strip()
        cleaned_list.append(cleaned_item)

    return cleaned_list

problem_standalone_components = clean_list(problem_standalone_components)

matrix_image = create_matrix_image(the_matrix, problem_standalone_components)
matrix_image.save("generic_complex_problem_matrix.png")
```

```python
# define the attitudes for each effect type
attitudes = {
    "deny": {"harmony": 0.05, "excitement": 0.05, "discipline": 0.90},
    "reinforce": {"harmony": 0.90, "excitement": 0.05, "discipline": 0.05},
    "circumvent": {"harmony": 0.30, "excitement": 0.40, "discipline": 0.30},
    "ignore": {"harmony": 0.20, "excitement": 0.60, "discipline": 0.20},
    "unclear": {"harmony": 0.33, "excitement": 0.33, "discipline": 0.33},
}

# function to compute overall attitude
def compute_predicted_attitude(row):
    source_purpose_effect_type = row["desired source purpose effect type"]
    source_capability_effect_type = row["desired source capability effect type"]
    sink_purpose_effect_type = row["predicted sink's purpose effect type"]
    sink_capability_reaction_effect_type = row["predicted sink's capability reaction effect type"]
    
    effect_types = [
        source_purpose_effect_type,
        source_capability_effect_type,
        sink_purpose_effect_type,
        sink_capability_reaction_effect_type,
    ]

    total_harmony = 0
    total_excitement = 0
    total_discipline = 0

    for effect_type in effect_types:
        total_harmony += attitudes[effect_type]["harmony"]
        total_excitement += attitudes[effect_type]["excitement"]
        total_discipline += attitudes[effect_type]["discipline"]

    avg_harmony = round(total_harmony / len(effect_types),2)
    avg_excitement = round(total_excitement / len(effect_types),2)
    avg_discipline = round(total_discipline / len(effect_types),2)
    
    attitude_tuple = (avg_harmony,avg_excitement, avg_discipline)
    
    if avg_harmony > avg_excitement + avg_discipline:
        attitude_type = "in Harmony"
    elif avg_discipline > avg_harmony + avg_excitement:
        attitude_type = "in Discipline"
    else:
        attitude_type = "in Excitement"
    

    return {
        "predicted attitude_type": attitude_type,
        "predicted attitude_tuple": attitude_tuple,
    }

# apply the attitude analysis for every source and sink
df_combinations["predicted attitude"] = df_combinations.apply(lambda row: compute_predicted_attitude(row)["predicted attitude_type"], axis=1)
# apply the attitude analysis for every source and sink and create a new column for attitude_tuple
df_combinations["predicted attitude tuple (h,e,d)"] = df_combinations.apply(lambda row: compute_predicted_attitude(row)["predicted attitude_tuple"], axis=1)
```
